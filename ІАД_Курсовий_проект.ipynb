# Завдання 6

Побудуйте модель регресії даних з файлу Sample99.csv за допомогою алгоритму SVR.

```python
path_to_6 = '/content/drive/MyDrive/ІАД Курсовий проект/Sample99.csv'
```

```python
data_6 = pd.read_csv(path_to_6)
data_6
```

```python
x = np.array(data_6.Param1).reshape(-1, 1)
y = np.array(data_6.Param2).reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,  y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.svm import LinearSVR
model = LinearSVR()
model.fit(x_train, y_train)
```

```python
pred = model.predict(np.array(data_6.Param1).reshape(-1,1))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data_6.Param1, data_6.Param2)
plt.plot(data_6.Param1, pred, 'red')
```

```python
from sklearn import metrics
```

```python
test_pred = model.predict(x_test)
```

```python
metrics.r2_score(y_test, test_pred)
```
